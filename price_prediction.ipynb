## Data Visualization (10 Marks)
13.	Create a Count Plot for Fuel Type Create a countplot showing the number of cars by fuel type.
- Title: "Count of Cars by Fuel Type"
- X-label: Fuel Type
- Y-label: Count (2 marks)


```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df_cleaned, x='Fuel type', palette='viridis', legend=False)
plt.title('Count of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()
```

14.	Create a Scatterplot of Engine Size vs Price Create a scatterplot showing the relationship between engine size and car price. Provide observations. (2 marks)

```python
# Convert to string, remove the word 'Turbo', and convert to a float
df['Engine volume'] = df['Engine volume'].astype(str).str.replace(' Turbo', '', case=False).astype(float)

# This removes that 26-million outlier so we can see the rest of the data
df_filtered = df[df['Price'] < 500000]

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='Engine volume', y='Price', alpha=0.6, color='b')

# Add labels and formatting
plt.title('Engine Volume vs Car Price (Outliers Removed)')
plt.xlabel('Engine Volume (Liters)')
plt.ylabel('Price')

# Optional: Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

15.	Create a Histogram of Car Prices Create a histogram of car prices. Explain the distribution. (2 marks)

```python
# Adjust this number up or down depending on what you want to focus on
df_normal_prices = df[df['Price'] < 100000]

# plot
plt.figure(figsize=(10, 6))
sns.histplot(df_normal_prices['Price'], bins=40, kde=True, color='blue') # type: ignore 

# Add labels and formatting
plt.title('Distribution of Car Prices (Under $100,000)')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')

# This forces the x-axis to use normal numbers instead of scientific notation
plt.ticklabel_format(style='plain', axis='x')

plt.show()
```

16.	Create a Pairplot for Selected Features Create a pairplot showing relationships between Price, Mileage, Engine Size, and Horsepower. (2 marks)

```python
# clean milage column
df['Mileage'] = df['Mileage'].astype(str).str.replace(' km', '', case=False).str.replace(',', '').astype(int)

# filter outliers 
df_filtered = df[df['Price'] < 500000]

# select features
features = ['Price', 'Mileage', 'Engine volume', 'Cylinders']

# Create plot
sns.pairplot(df_filtered[features])
plt.show()
```

17.	Create a Boxplot to Detect Outliers Create boxplots for numerical variables to identify possible outliers. Explain your observations. (2 marks)

```python
# Select the meaningful numerical features
features_to_plot = ['Price', 'Prod. year', 'Cylinders', 'Airbags'] 

# Set up a wide figure to hold multiple subplots side-by-side
plt.figure(figsize=(16, 5))

# Create a separate boxplot for each feature
for i, col in enumerate(features_to_plot, 1):
    plt.subplot(1, len(features_to_plot), i)
    sns.boxplot(y=df_filtered[col], color='skyblue') 
    plt.title(f'Boxplot of {col}')
    plt.ylabel('') 

plt.tight_layout()
plt.show()
```
